```python
url2 = ('http://newsapi.org/v2/top-headlines?'
       'from=2020-07-18&'
       'country=au&'
       'language=en&'
       'pageSize=100&'
       'sortBy=popularity&'
       'apiKey=7cbc46d369ec49efba7d4e1bd3df8b1a')

response2 = requests.get(url2)
```

```python
def print_title_description(response):
    i = 0
    while i < len(response.json()['articles']):
        #article_dict['content']
        article_dict = response1.json()['articles'][i]
        text_to_be_analysed = article_dict['title'] + ". " +article_dict['description']
        
        blob = TextBlob(text_to_be_analysed)
        total_score = 0
        str_score = ''
        num_sentence = 0
        for sentence in blob.sentences:
            total_score += sentence.sentiment.polarity
            str_score = str_score + ' ' + str(sentence.sentiment.polarity)
            num_sentence += 1
        average = total_score/num_sentence
        print(text_to_be_analysed)
        print("Score :"+str_score)
        print("Total Score: " + str(total_score))
        print("Average: " + str(average))
        print("")
        i += 1
```

```python
def print_title_description_combined(response):
    i = 0
    max_score = 0
    str_save = ""
    while i < len(response.json()['articles']):
        #article_dict['content']
        article_dict = response1.json()['articles'][i]
        text_to_be_analysed = article_dict['title'] + ". " +article_dict['description']
        
        blob = TextBlob(text_to_be_analysed)
        
        score = blob.sentiment.polarity
        if (max_score < score):
            max_score = score
            str_save = text_to_be_analysed
        if score > 0.5:
        
            print(text_to_be_analysed)
            print("Score: " + str(score))
            print("")
        i += 1
    return str_save, max_score
```

```python
str_save, max_score = print_title_description_combined(response3)
```
